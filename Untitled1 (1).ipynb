```python
import numpy as np
import random
import statistics 
import pandas as pd
import matplotlib.pyplot as plt
import datetime 
from decimal import *
import dask
```

```python
df2 = pd.read_csv(r'/optionspaperprices.csv')
df2.drop(df2.columns[0], axis=1, inplace=True)
```

```python
df2 = df2.ffill()
df2.isna().sum()

#check for nan values 
```

```python
df2.replace(0,4663, inplace=True)
#4663
```

```python
df2['cumsum'] = df2['futuresprice'].pct_change().cumsum()
df2.tail()
#get cumulative percent change 
```

```python
def my_function(x):
    if x['cumsum'] >= 0.01 :
        return 'long'
    if x['cumsum'] <= -0.01:
        return 'short'
    else:
        return 'do nothing'
# create signal
```

```python
df2['signal'] = df2.apply(my_function, axis=1)
#apply signal
```

```python
signaldf = df2.loc[df2['signal'] != 'do nothing']
signaldf.head()
```

```python
#Assume slippage of 2mins 
#Assume trading cost of 2%
#Leverage ratio of 2
cost = 2
slippage = 1200
leverage = 2
shortreturns = [] 
pct_returns = []
for i in range(len(df2['cashprice'])):
  if df2['signal'][i] == 'short':
    short = (df2['cashprice'][i+slippage+1333] - df2['cashprice'][i+slippage])*leverage*1-cost/100
    short_pct = ((df2['cashprice'][i+slippage+1333]*leverage*1+cost/100) - (df2['cashprice'][i+slippage]*leverage*1-cost/100))/(df2['cashprice'][i+slippage]*leverage*1-cost/100)
    shortreturns.append(short)
    pct_returns.append(short_pct)


```

```python
cost = 2
slippage = 1200
leverage = 2
longreturns = [] 
for j in range(len(df2['cashprice'])):
  if df2['signal'][j] == 'long':
    buy = (df2['cashprice'][j+slippage+1333] - df2['cashprice'][j+slippage])*leverage*1-cost/100

    longreturns.append(buy)
```

```python
df2['signal'].unique()
# No buy signals
```

```python
for k in range(len(pct_returns)):
  pct_returns[k] = pct_returns[k]*100
#convert to percentages 
```

```python
cumulative_returns = []
f=0
for g in range(0,len(pct_returns)):
    f+=pct_returns[g]
    cumulative_returns.append(f)
#calculate cumulative return in %
```

```python
plt.plot(np.arange(1,190,1),cumulative_returns)
plt.xlabel('N trades')
plt.ylabel('percentage return')
plt.title('Strategy returns')
```

```python
openprices = []
closeprices = []
cost = 2
slippage = 1200
for v in range(len(df2['cashprice'])):
  if df2['signal'][v] == 'short':
    open = df2['cashprice'][v+slippage]*1-cost/100   ###sell
    close = df2['cashprice'][v+1333+slippage]*1+cost/100 ###buy
    openprices.append(open)
    closeprices.append(close)
  prices = closeprices + openprices
```

```python


prices = openprices
returns = pct_returns

last_price = prices[-1]

#Number of Simulations
num_simulations = 1000
num_days = len(openprices)

simulation_df_open = pd.DataFrame()

for x in range(num_simulations):
    count = 0
    daily_vol = np.std(pct_returns)
    
    price_series = []
    
    price = last_price * (1 + np.random.normal(0, daily_vol))
    price_series.append(price)
    
    for y in range(num_days):
        if count == num_days - 1:
            break
        price = price_series[count] * (1 + np.random.normal(0, daily_vol))
        price_series.append(price)
        count += 1
    
    simulation_df_open[x] = price_series
```

```python
prices = closeprices
returns = pct_returns

last_price = prices[-1]

#Number of Simulations
num_simulations = 1000
num_days = len(openprices)

simulation_df_close = pd.DataFrame()

for x in range(num_simulations):
    count = 0
    daily_vol = np.std(pct_returns)
    
    price_series = []
    
    price = last_price * (1 + np.random.normal(0, daily_vol))
    price_series.append(price)
    
    for y in range(num_days):
        if count == num_days - 1:
            break
        price = price_series[count] * (1 + np.random.normal(0, daily_vol))
        price_series.append(price)
        count += 1
    
    simulation_df_close[x] = price_series
```

```python
open = simulation_df_open
close = simulation_df_close
```
