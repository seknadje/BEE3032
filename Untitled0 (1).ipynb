```python
#Loading SPY ETF price data of whole day of 10/11/21
from dask import dataframe as df1
from dask.array import from_array as fa
path = "/content/drive/MyDrive/cashspy.csv"
cash = df1.read_csv(path)
cash.head()  
#Load and read data using parrallel programming library dask as dataset 4gb

```

```python
cash['Timestamp'] = cash.Timestamp.str.replace(':' , '')
cash['Timestamp'] = cash.Timestamp.str.replace('.' , '')
cash.head()
#Remove string characters of : and . to enable float datatype & computation

```

```python
cash['Timestamp']=cash['Timestamp'].astype(float)
cash['Timestamp'] = cash['Timestamp']/10**9
cash.head()
# Put milliseconds as a decimal after hh/mm/ss/milliseconds
```

```python
cash['Timestamp'] = cash['Timestamp'].round(1)
cash.head()
#Round milliseconds to nearest millisecond
```

```python
#E-minis SPY futures maturity in Dec 2021 on same day as SPY
path2 = "/content/drive/MyDrive/ESZ1.csv"
futures = df1.read_csv(path2,decimal = '.')
futures.head()
#Load and read data using parrallel programming library dask as dataset 4gb
```

```python
futures['LocalTime'] = futures['LocalTime']/10**9
futures.head()
# Put milliseconds as a decimal after hh/mm/ss/milliseconds
```

```python
futures['LocalTime'] = futures['LocalTime'].round(1)
futures.head()
#Round milliseconds to nearest millisecond
```

```python
cash = cash[cash['Price']>0.0]
cash.head()
#Drop cash rows where price = 0 (market not open)
```

```python
futures = futures[futures['LocalDate']>20211109]
futures.head()
#Remove all data from the day before
```

```python
futures['LocalTime'] = futures['LocalTime'] + 10**4
futures.head()
#Convert CST time to EST time  EST = CST + 1h
```

```python
futures = futures[futures['LocalTime']>=40000.0]
futures.head()
#Get data starting at 4am onwards as that is when SPY opens
```

```python
cash = cash[cash['Timestamp']<=185960.0]
cash.tail()
#Remove for time > 19h as futures not open then
```

```python
cash.columns = ['DateCash','Timestampcash','EventTypecash','Tickercash','Pricecash','Quantitycash','Exchangecash','Conditionscash']
cash.head()
#Renaming columns cash
```

```python
futures.columns = ['UTCDatefutures',	'UTCTimefutures',	'LocalDatefutures',	'LocalTimefutures',	'Tickerfutures',	'SecurityIDfutures',	'TypeMaskfutures',	'Typefutures',	'Pricefutures',	'Quantityfutures',	'Ordersfutures',	'Flagsfutures']
futures.head()
#Renaming columns futures
```

```python
futures = futures.drop(columns=['UTCDatefutures','UTCTimefutures','SecurityIDfutures',	'TypeMaskfutures','Flagsfutures'])
futures.head() 
#Droppping columns that I dont need 
```

```python
cash = cash.drop(columns=['Conditionscash'])
cash.head() 
#Droppping columns that I dont need 
```

```python
cash = cash.set_index('Timestampcash')

median_fun = df1.Aggregation(name="median",chunk=lambda s: s.median(),agg=lambda s0: s0.sum(),)

cash = cash.groupby("Timestampcash")["Pricecash"].agg(median_fun)
cash.head()
#Create mid function median_fun for parrallel programming
# this computes the median on each partition for mid prices
#this combines results across partitions; 
#the input should just be a list of length 1
# output is mid prices for everymilisecond
```

```python
futures= futures.set_index('LocalTimefutures')

median_fun = df1.Aggregation(name="median",chunk=lambda s: s.median(),agg=lambda s0: s0.sum(),)

futures = futures.groupby("LocalTimefutures")["Pricefutures"].agg(median_fun)
futures.head()
```

```python
cash = cash.to_frame()
futures = futures.to_frame()
df = futures.merge(cash, how='outer',left_index = True,right_index=True)
df.head()
#merge datasets on same time interval
```

```python
df['Pricecash'] =  df['Pricecash'].fillna(method="ffill")
df.head()
#Fill nan values for cash with previous value as price has not changed
```

```python
df = df.diff(axis=0)
df = df.dropna()
df.head()
#calculate 1st difference for stationarity + drop nan values
```

```python
futureslist=df['Pricefutures'].compute().tolist() 
```

```python
cashlist=df['Pricecash'].compute().tolist() 
```

```python
from statsmodels.tsa.stattools import adfuller
res = adfuller(df['Pricecash'])
res2 = adfuller(df['Pricefutures'])
 
# Printing the statistical result of the adfuller test
print('Augmneted Dickey_fuller Statistic cash: %f' % res[0])
print('p-value cash : %f' % res[1])
print('Augmneted Dickey_fuller Statistic futures: %f' % res2[0])
print('p-value futures : %f' % res2[1])

#ADF < critical values so time series stationary
# Pvalue < 0.05 level so time series stationary

```

```python
def shift(xs, n):
    e = np.empty_like(xs)
    if n >= 0:
        e[:n] = np.nan
        e[n:] = xs[:-n]
    else:
        e[n:] = np.nan
        e[:n] = xs[-n:]
    return e
```

```python
dt = pd.DataFrame(list(zip(cashlist, futureslist)),
               columns =['cashprice', 'futuresprice'])
dt.head()
```

```python
l = []
for i in range(-6000,6000):
  correlation = dt['cashprice'].shift(i).corr(dt['futuresprice'])
  l.append(correlation)
# calculate corrells for every shift 
```

```python
l[0:5]
```

```python
t = []
for j in range(-6000,6000):
  t.append(j)
t[0:5]
```

```python
for k in range(len(t)):
  t[k] = t[k]/10
t[0:5]
```

```python
l_graph = []
for r in range(len(l)):
  normalized = (l[r] - min(l))/(max(l)-min(l))
  l_graph.append(normalized)


%matplotlib inline

fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot(111)
ax.set_xlabel('Time Step in Seconds range = -+10mins with 0.1 sec interval')
ax.set_ylabel('Normalized correllation to adjust for scale difference 0.5 means correl = 0')
ax.set_title('Cash vs futures')
xs = np.linspace(-600, 600, len(t))
ys = l_graph
ax.plot(xs, ys, marker='.')

#Plot of cash vs futures correll vs timestep
# Stat arb possible because correl > 0 at a time step that is not zero
#When cash has a shift forward of 180 seconds the correlation is highest 
#which means futures at t = 0 can predict cash at t = approx 180
# when futures moves by 1 cash moves by 0.095 2.533 mins later
```
